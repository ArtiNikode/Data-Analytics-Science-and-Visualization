```python
import matplotlib.pyplot as mpl
import pandas

df=pandas.read_csv('Company.csv')
df1=df[['Company','Average Salary']]

mpl.plot(df1['Company'],df1['Average Salary'],color='blue')
mpl.scatter(df1['Company'],df1['Average Salary'],color='red')

mpl.show()

```

```python
import matplotlib.pyplot as mpl
import pandas

df=pandas.read_csv('Company.csv')
df2=df[['Company','Jobs']]

mpl.bar(df2['Company'],df2['Jobs'],color='blue')
mpl.xlabel('Company')
mpl.ylabel('Number of jobs')
mpl.show()
```

```python
import matplotlib.pyplot as mpl
import pandas

df=pandas.read_csv('Company.csv')
df3=df[['Jobs']]

mpl.pie(df3['Jobs'])
mpl.title('Number of Jobs')

mpl.show()
```

```python
import matplotlib.pyplot as mpl
import pandas

df=pandas.read_csv('Company.csv')
df4=df[['Company','Jobs']]
df5=df[['Company','Average Salary']]

fig, axes=mpl.subplots(1,2)
axes[0].plot(df4['Company'],df4['Jobs'],color='purple')
axes[1].plot(df5['Company'],df5['Average Salary'],color='green')

mpl.show()             
```
